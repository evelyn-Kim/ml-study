```python
import pandas as pd
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

path = 'C:/Users/pc/OneDrive/Desktop/hyowon/ml-study/week2/2020.xlsx'
df = pd.read_excel(path, skiprows=1)
feature = df[["행정구역별(읍면동)","총인구 (명)"]]
```

```python
def top5graph(df, feature):
    path = 'C:/Users/pc/OneDrive/Desktop/hyowon/ml-study/week2/2020.xlsx'
    df = pd.read_excel(path, skiprows=1)
    feature = df[["행정구역별(읍면동)","총인구 (명)"]]
    df = df.loc[5:29]
    df = df.nlargest(5, "총인구 (명)")
    df = df.plot.bar(x="행정구역별(읍면동)", y= "총인구 (명)", rot=0, title = "서울특별시 상위 5개의 지역")
    return df

top5graph(df, feature)
```

```python
def circlegraph(df, feature):
    path = 'C:/Users/pc/OneDrive/Desktop/hyowon/ml-study/week2/2020.xlsx'
    df = pd.read_excel(path, skiprows=1)
    feature = df[["행정구역별(읍면동)","총인구 (명)"]]
    df = df.loc[5:29]
    a = df["총인구 (명)"].tail(20)
    s = a.sum(axis=0)
    df = df.nlargest(5, "총인구 (명)")
    df = df.reset_index(drop=True)
    df = df.append({"행정구역별(읍면동)": "         기타 ","총인구 (명)": s },ignore_index=True)
    df = df.plot.pie(figsize=(9,8), y = "총인구 (명)", labels = df["행정구역별(읍면동)"], title = "서울특별시 상위 5개의 지역")
    return df

circlegraph(df, feature)
```
